```python
from PIL import ImagePath
import glob
from PIL import Image, ImageOps

img = Image.open(fullimagepath) 
raw_img = Image.open(rawimagepath) 
img = ImageOps.grayscale(img)
raw_img = ImageOps.grayscale(raw_img)

boundingWidth, boundingHeight = img.size
width, height = raw_img.size

for i in range(width-boundingWidth):
    for j in range(height-boundingHeight):
        noMatch = False
        #print(i , j)
        k = min(boundingHeight,boundingWidth)

        for z in range(k):
            if(raw_img.getpixel((i+z,j+z)) != img.getpixel((z,z))):
                noMatch = True
                #print("no match ", i,j)
                break
        if(noMatch == False):
            print(i,j)
            print(i+boundingWidth, j + boundingHeight)

```
