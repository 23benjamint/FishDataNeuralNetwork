```python
import os
from torchvision.transforms import ToPILImage, Resize
from IPython.display import Image
from torch.utils.data import DataLoader
import torch
from PIL import Image, ImageOps


x=0 
counter = 0 
pathraw= "/Users/benjamintian/Documents/Fish_Data/images/raw_images"


for root, dirs, files in os.walk("/Users/benjamintian/Documents/Fish_Data/images/cropped", topdown=False):
   for name in files:
      image_filepath = os.path.join(root, name)
      #print(image_filepath)
      rawimage_filepath= os.path.join(pathraw,name)
      rawimage_filepath = rawimage_filepath.replace(".png", ".jpg")
      print(rawimage_filepath)

      if (counter >10 ):
         break
      counter +=1
      
      label = image_filepath.split('/')[-1]
      label = label.split("_")
      label = label[0] +" "+ label[1]

      cropped_img = Image.open(image_filepath) 
      raw_img = Image.open(rawimage_filepath) 

      cropped_img = ImageOps.grayscale(cropped_img)
      raw_img = ImageOps.grayscale(raw_img)

      boundingWidth, boundingHeight = cropped_img.size
      width, height = raw_img.size

      for i in range(width-boundingWidth):
         for j in range(height-boundingHeight):
            noMatch = False
            k = min(boundingHeight,boundingWidth)
            for z in range(k):
               if(raw_img.getpixel((i+z,j+z)) != cropped_img.getpixel((z,z))):
                     noMatch = True
                     break
               if(noMatch == False):
                  #print(i,j)
                  #print(i+boundingWidth, j + boundingHeight)

                  y = names.index(label)
                  f= open(f"/Users/benjamintian/Documents/datasets/fishdata/labels/fishdata/train/train{x}.txt", mode= 'w+')

                  centerwidth = ((i + i +boundingWidth) /2)/width
                  centerheight =  ((j + j + boundingHeight )/2)/height
                  outputwidth = boundingWidth/width
                  outputheight = boundingHeight/width

                  output = str(y) + " " +  str(centerwidth) + " " +  str(centerheight) + " " + str(outputwidth) +" "+ str(outputheight)
                  #print(image_filepath)
                  #print(output)
                  f.write(output)
                  f.close()
      x = x + 1
      
   for name in dirs:
      print(os.path.join(root, name))
```
